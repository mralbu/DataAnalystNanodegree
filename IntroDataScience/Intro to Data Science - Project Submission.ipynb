```python
stations = df[['station', 'latitude', 'longitude']].drop_duplicates()
stations = stations.groupby('station').agg({'latitude':np.mean, 'longitude':np.mean}).reset_index(level=0)
stations.columns = ['station', 'mean_lat', 'mean_lon']
stations.head()
```

```python
coords = geopandas.GeoSeries([Point(lon, lat) for lat, lon in zip(stations.mean_lat, stations.mean_lon)])
stations = geopandas.GeoDataFrame(stations, geometry=coords)
```

```python
def neighborhood(point):
    for i, nb in nybb.iterrows():
        if point.within(nb['geometry']):
            return nb['BoroName']
```

```python
stations['neighbourhood'] = stations.geometry.apply(neighborhood)
stations.head()
```

```python
df = geopandas.GeoDataFrame(pd.merge(df, stations, how='left', on='station'))
```

### Section 1. Statistical Test

**Question**

Which Statistical test did you use to analyse the NYC subway data? Did you use a one-tail or a two-tail P value?

*Answer* <br>


I used the Mann-Whitney test, a non-parametric test, as the ENTRIESn_hourly variable is not normally distributed. I used a two-sided test, to evaluate if the hourly entries on rainy days and non-rainy days present distributions with significantly different ranks, that is, if these distributions are significantly different.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

sns.boxplot(df.ENTRIESn_hourly, groupby=df.rain, ax=ax1)
sns.boxplot(df.ENTRIESn_hourly, groupby=df.rain, ax=ax2)
ax2.set_ylim(0, 2600)
ax2.set_title('Zoomed In')

fig.tight_layout()
```

```python
import scipy.stats

with_rain = df[df.rain == 1].ENTRIESn_hourly
without_rain = df[df.rain == 0].ENTRIESn_hourly

U, p = scipy.stats.mannwhitneyu(with_rain, without_rain)

print('The Mann-Whitney statistic is: {0}, p-value {1}'.format(U, 2*p))
```

The Mann-Whitney statistic two-sided p-value is lower than 0.05, therefore,  we reject the hypothesis that there is no difference between the ranks of the rainy and non rainy days hourly Entries distributions, with a confidence of 95%.

**Question**

Why is this statistical test appropriate or applicable to the dataset?

*Answer* <br>


The ENTRIESn_hourly variable is not normally distributed. Because of that, a Welch's T-Test could not be used, as it assumes normally distributed variables. <br>
The Mann-Whitney test is a nonparametric test and does not assume normal data, and is, therefore more appropriate.

```python
g = sns.FacetGrid(df, hue='rain', col='rain')
g.map(sns.kdeplot,'ENTRIESn_hourly', shade=True)
g.fig.suptitle('Hourly Entries Prob Dens')
g.fig.tight_layout()
g.set_xticklabels(rotation=30);
```

**Question**

What results did you get from this statistical test? These should include the numerical values: p-values, as well as the means for each of the two samples under test.

*Answer* <br>


```python
print('Rainy Days mean hourly Entries: {}'.format(with_rain.mean()))
print('Non-Rainy Days mean hourly Entries: {}'.format(without_rain.mean()))
print('')
print('Rainy Days median hourly Entries: {}'.format(with_rain.median()))
print('Non-Rainy Days median hourly Entries: {}'.format(without_rain.median()))
print('')
print('Mann-Whitney Statistic: {}'.format(U))
print('Two-Sided p-value: {}'.format(2*p))
print('')
```

**Question**

What is the significance of these results?

*Answer* <br>


These results lead us to reject the null-hypothesis that these distributions have equal medians, with a confidence of 95% (this confidence setting was chosen for this test, considering a p-value threshold of 0.05). <br>
As the one-sided test p-value can be shown as well to be bellow our 0.05 threshold, we conclude that people are more likely to take the subway on rainy days.

What approach did you use to compute the coefficients theta and produce prediction in your regression model?

I used OLS (statsmodels).

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols, glm
```

```python
lm_model = ols('ENTRIESn_hourly ~ C(UNIT) + C(hour) + C(rain) + C(day_week)', df)
reg_lm = lm_model.fit()
reg_lm.summary()
```

We can also perfom an Analysis of Variance comparing linear models with and without the rain factor.

```python
with_rain_lmmodel = ols('ENTRIESn_hourly ~ C(UNIT) + C(hour) + C(rain) + C(day_week)', df).fit()
no_rain_lmmodel = ols('ENTRIESn_hourly ~ C(UNIT) + C(hour) + C(day_week)', df).fit()

print 'Rain Coefficient P-Value: {0:.6g}'.format(with_rain_lmmodel.pvalues['C(rain)[T.1]'])

sm.stats.anova_lm(no_rain_lmmodel, with_rain_lmmodel)
```

Assuming a significance level of 95%, we would be inclined to reject the hypothesis that the rain factor is significant in our model.
<br>
However, a gaussian linear model is not adequate for our dataset, as it violates the assumption of normally distributed residuals, as can be seen by the high kurtosis skewed residuals histogram graphed below.
<br>
The model's summary function presents, as well, the results of two residuals normality tests, Omnibus and Jarque-Bera, that also atest the non-normality of the model's residuals.

```python
fig, ax = plt.subplots();
ax.hist(reg_lm.resid, bins=30, normed=True)
ax.set_ylabel('Normalized Residuals Counts')
ax.set_title('Residuals Histogram');
```

Generalized linear models were developed as a flexible generalization for linear models with response variables with non-normal errors.

$$
g(\mu_i) = \beta_0 + \beta_1x_{1i} + ... + \beta_px_{pi} \\\
E(Y) = \mu \\\
Y \sim ExponentialFamily(.)
$$

Where $g(.)$ is called the link function, and Y is a response variable with errors distributed by a [Exponential Family](http://en.wikipedia.org/wiki/Exponential_family) probability density.

[Poisson Regression](http://en.wikipedia.org/wiki/Poisson_regression) is a classical model for count data. It assumes the response variable has a poisson distribution and that the logarithm of it's expected value can be modeled with linear parameters. A poisson distribution has equal mean and variance. We can see that this is not true for our dataset, and the model is said to be overdispersed. A good candidate for overdispersed count models is the [Negative Binomial model](http://www.ats.ucla.edu/stat/stata/dae/nbreg.htm).

```python
print 'Mean Hourly Entries: {}'.format(df.ENTRIESn_hourly.mean().round(1))
print 'Variance of Hourly Entries: {}'.format(df.ENTRIESn_hourly.var().round(1))
```

Fitting these models, we can see that models with more adequate assumptions have lower deviance.

```python
#OLS linear model
glm_gaussian_model = glm('ENTRIESn_hourly ~ C(neighbourhood) + C(hour) + C(rain) + C(day_week)', df, family=sm.families.Gaussian())
reg_glm_gaussian = glm_gaussian_model.fit(method='bfgs')

#GLM Gaussian model with log Link Function
glm_gaussianLog_model = glm('ENTRIESn_hourly ~ C(neighbourhood) + C(hour) + C(rain) + C(day_week)', df, family=sm.families.Gaussian(link=sm.families.links.log))
reg_glm_gaussianLog = glm_gaussianLog_model.fit(method='bfgs')

#GLM Poisson model with default log Link Function
glm_Poisson_model = glm('ENTRIESn_hourly ~ C(neighbourhood) + C(hour) + C(rain) + C(day_week)', df, family=sm.families.Poisson())
reg_glm_Poisson = glm_Poisson_model.fit(method='bfgs')

#GLM Negative Binomial model with default log Link Function
glm_NegBinomial_model = glm('ENTRIESn_hourly ~ C(neighbourhood) + C(hour) + C(rain) + C(day_week)', df, family=sm.families.NegativeBinomial())
reg_glm_NegBinomial = glm_NegBinomial_model.fit(method='bfgs')

model_deviance_list = np.array([reg_glm_gaussian.deviance, reg_glm_gaussianLog.deviance, reg_glm_Poisson.deviance, reg_glm_NegBinomial.deviance])
```

```python
ax = sns.barplot(np.arange(4), y=model_deviance_list)
ax.set_yscale('log')
ax.set_title('Model Deviance')
ax.set_xticklabels(['Gaussian', 'Log Gaussian', 'Poisson', 'Negative Binomial']);
```

```python
with sns.plotting_context('notebook', font_scale=1):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
    
    ax1.set_title('Gaussian Model Deviance Residuals\nDeviance: '+'{0:.2g}'.format(reg_glm_gaussian.deviance))
    ax1.plot(reg_glm_gaussian.resid_deviance.index, reg_glm_gaussian.resid_deviance, 'b.')
    
    ax2.set_title('Log-Gaussian Model Deviance Residuals\nDeviance: '+'{0:.2g}'.format(reg_glm_gaussianLog.deviance))
    ax2.plot(reg_glm_gaussianLog.resid_deviance, 'b.')
    
    ax3.set_title('Poisson Model Deviance Residuals\nDeviance: '+'{0:.2g}'.format(reg_glm_Poisson.deviance))
    ax3.plot(reg_glm_Poisson.resid_deviance, 'b.')
    
    ax4.set_title('Negative Binomial Model Deviance Residuals\nDeviance: '+'{0:.2g}'.format(reg_glm_NegBinomial.deviance))
    ax4.plot(reg_glm_NegBinomial.resid_deviance.index, reg_glm_NegBinomial.resid_deviance, 'b.')

    ax2.set_ylim(ax1.get_ylim())
    ax4.set_ylim(ax3.get_ylim())
    
    ax3.set_xticklabels(reg_glm_Poisson.resid_deviance.index)
    ax4.set_xticklabels(reg_glm_NegBinomial.resid_deviance.index)
    
    fig.tight_layout()
```

```python
print 'Poisson Regression: Rain Coefficient P-Value: {0:.3g}'.format(reg_glm_Poisson.pvalues['C(rain)[T.1]'])
reg_glm_Poisson.summary()
```

```python
print 'Negative Binomial Regression: Rain Coefficient P-Value: {0:.3g}'.format(reg_glm_NegBinomial.pvalues['C(rain)[T.1]'])
reg_glm_NegBinomial.summary()
```

The pvalues for the rain coefficient in the Poisson and Negative Binomial Regression models lead to different conclusions: for the poisson model, we reject the null hypothesis that the rain coefficient is not significant; for the Negative Binomial Regression we do not reject the null hypothesis that the rain coefficient is not significant. 
<br>
As we can tell by this analysis, the assumptions used in our models have paramount importance in Statistical Hypothesis Testing analysis.
